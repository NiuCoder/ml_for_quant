```python
import datetime
import numpy as np
import pandas as pd
import pandas_datareader.data as web
from pypinyin import pinyin, Style
```

```python

#此段代码已经执行完毕，并生成了优化后的沪深300csv文件
#为了从yahoo抓数据，需要对沪深300股票做处理
hs300s = pd.read_csv('hs300.csv',index_col=0)
#对code依次补0，并且添加新的一列symbol存放股票名称的拼音缩写，根据沪深代码规则添加后缀，最后把新的dataframe保存到csv文件
codes = []
abbr = []
symbols = []
for code in hs300s['code']:
    code = str(code).zfill(6)
    codes.append(code)
    if(code[:2] != '60'):
        symbols.append(code+'.sz')
    else:
        symbols.append(code+'.ss')
for name in hs300s['name']:
    py_letters = pinyin(unicode(name,'utf8'),style=Style.FIRST_LETTER)
    name = ''
    for le in py_letters:
        name = name+le[0]
    abbr.append(name.replace(' ',''))
hs300s['code']=codes
hs300s['abbr'] = abbr
hs300s['symbol'] = symbols
hs300s.to_csv('hs300s.csv',encoding='utf-8',index=False)
```

```python
# 计算一只股票的hurst指数，如果小于0.5表明是均值回归的，如果大于0.5表明是趋势的，等于0.5是所谓的几何布朗运动

from numpy import cumsum, log, polyfit, sqrt, std, subtract
from numpy.random import randn

def hurst(ts):

    """Returns the Hurst Exponent of the time series vector ts"""
    # Create the range of lag values
    lags = range(2, 100)

    # Calculate the array of the variances of the lagged differences
    tau = [sqrt(std(subtract(ts[lag:], ts[:-lag]))) for lag in lags]

    # Use a linear fit to estimate the Hurst Exponent
    poly = polyfit(log(lags), log(tau), 1)

    # Return the Hurst exponent from the polyfit output
    return poly[0]*2.0 
#构建布朗运动，均值回归和趋势序列，计算其赫斯特指数用于对比
gbm = log(cumsum(randn(100000))+1000)
mr = log(randn(100000)+1000)
tr = log(cumsum(randn(100000)+1)+1000)
hurst_zzxq = hurst(zxzq['Adj Close'])
hurst_wk = hurst(wk['Adj Close'])

print 'Hurst(GBM): %s' % hurst(gbm)
print 'Hurst(MR): %s' % hurst(mr)
print 'Hurst(TR): %s' % hurst(tr)
print 'Hurst(中信证券): %s' % hurst_zzxq
print 'Hurst(万科A): %s' % hurst_wk 
```
